```python
import pandas as pd, numpy as np, seaborn as sns, random
import matplotlib.pyplot as plt
import matplotlib.colors as clr
from sklearn.neural_network import MLPRegressor, MLPClassifier


df = pd.read_csv('todos_os_mapas.csv')
df.drop(columns = ['Unnamed: 0'], inplace=True)
```

```python
def separador(df):
    # As partidas estão separadas por linhas "NaN", isto cria uma series dizendo se a linha separa uma partida ou não
    temp = df.isnull().apply(lambda x: all(x), axis=1)

    # # recupera os índices das linhas que separam as partidas para usar em laços
    separa = temp[temp == True].index.tolist()
    separa = [-1] + separa               # garante que pega a primeira partida se a primeira linha não é NaN
    #separa = separa + [len(temp)]       # garante que pega a  última  partida se a  última  linha não é NaN
    return separa
```

```python
separa = separador(df)
```

```python
# # separa o nome do time do nome do jogador, já que farei iterações nos times

temp1 = []
temp2 = []
for index, jogador in df[['Player-name']].iterrows():
    if type(df['Player-name'][index]) != type(1.0) and len(df['Player-name'][index].split()) != 0:
        nome = df['Player-name'][index].split()
        temp1.append(nome[-1])
        temp2.append(''.join(nome[:-1]))
    else:
        temp1.append(np.nan)
        temp2.append(np.nan)
df = df.drop(columns = ['Player-name'])
df = pd.concat([df, pd.DataFrame(temp1)], axis = 1)
df = pd.concat([df, pd.DataFrame(temp2)], axis = 1)

cols = df.columns.tolist()
cols[-2] = 'Team'
cols[-1] = 'Player-name'
df.columns = cols
df
```

```python
# # correlações dos atributos
plt.figure(figsize = (16, 16))
ax = sns.heatmap(df.corr(min_periods = 20),  cmap = 'cividis', linewidths = 0.003, linecolor = 'lightcyan')
ax.xaxis.tick_top()
ax.tick_params(axis = 'x', labelrotation = 90)
plt.show()
```

```python
## cria uma lista com os times do df que possuem mais de n mapas jogados
n = 10 ## mínimo 10, minha rede neural atual é com n = 49

teste = df[['Team', 'Player-name']]
teste = teste.groupby('Team').count() > n
teste = teste[teste['Player-name']]
teste = teste.index.tolist()

teste.remove('VKS') ## tem dois times com sigla VKS no dataframe: Team Vikings e VivoKeydStars

# # cria um dicionario onde as keys são os times e os itens são listas
# # com os índices do separador que possuem partidas do respectivo time 

jogos_por_time = {}
for team in teste:

    jogos_time = []
    temp = df[df['Team'] == team]
    i = temp.index[0]//11
    while (i<len(separa)-1) :        
        a = (np.array(temp.index.tolist()) > separa[i]) & (np.array(temp.index.tolist()) < separa[i+1])
        #a = set(range(separa[i]+1,separa[i+1]))
        #a = a.intersection(np.array(temp.index.tolist()))
        if len(temp[a]) == 5:
            jogos_time.append(i)    
        i+=1        
    jogos_por_time[team] = jogos_time
    
```

```python
# # atributos que vou usar
df2 = df[['ADR', 'KAST', 'R']]
df2
```

```python
# # montando os dataframes para uso da rede neural
# # sim, está ineficiente
treino_total = pd.DataFrame()
objetivo_total = pd.DataFrame()

for team in teste:
    indices = jogos_por_time[team]
    
    for i in range(len(indices[10:])):
        
        temp = df.loc[separa[indices[i+10]]+1:separa[indices[i+10]+1]-1, ['Team', 'Resultado']]
        
        if len(set(temp['Team'])) == 2:
            if list(temp['Team']).index(team) < 4:
                time1 = True
                other_team = list(temp['Team'])[5]
            else:
                other_team = list(temp['Team'])[0]
                time1 = False
                
        try:
            other_indices = jogos_por_time[other_team]
            other_i = other_indices.index(indices[i+10])
        except:
            other_i = 0
            pass
        if other_i >= 10:
            objetivo = df.loc[temp.index[0],'Resultado']            
            treino_team = pd.DataFrame()
            treino_other = pd.DataFrame()
            df_team = df2[['R', 'KAST', 'ADR']][df['Team'] == team]
            df_other = df2[['R', 'KAST', 'ADR']][df['Team'] == other_team]
            
            for j in range(10):
                treino_team = pd.concat([
                    treino_team,
                    df_team.loc[separa[indices[i+j]]+1:separa[indices[i+j]+1]-1, :].sort_values(by = 'R')['R']
                    ], axis = 0, ignore_index = True)
                treino_team = pd.concat([
                    treino_team,
                    df_team.loc[separa[indices[i+j]]+1:separa[indices[i+j]+1]-1, :].sort_values(by = 'R')['KAST']
                    ], axis = 0, ignore_index = True)
                treino_team = pd.concat([
                    treino_team,
                    df_team.loc[separa[indices[i+j]]+1:separa[indices[i+j]+1]-1, :].sort_values(by = 'R')['ADR']
                    ], axis = 0, ignore_index = True)
                treino_other = pd.concat([
                    treino_other,
                    df_other.loc[separa[other_indices[other_i-10+j]]+1:separa[other_indices[other_i-10+j]+1]-1, :].sort_values(by = 'R')['R']
                    ], axis = 0, ignore_index = True)
                treino_other = pd.concat([
                    treino_other,
                    df_other.loc[separa[other_indices[other_i-10+j]]+1:separa[other_indices[other_i-10+j]+1]-1, :].sort_values(by = 'R')['KAST']
                    ], axis = 0, ignore_index = True)
                treino_other = pd.concat([
                    treino_other,
                    df_other.loc[separa[other_indices[other_i-10+j]]+1:separa[other_indices[other_i-10+j]+1]-1, :].sort_values(by = 'R')['ADR']
                    ], axis = 0, ignore_index = True)
            if time1:
                treino = pd.concat( [treino_team, treino_other], axis = 0, ignore_index = True)
            else:
                treino = pd.concat( [treino_other, treino_team], axis = 0, ignore_index = True)
            
            treino_total = pd.concat([
                treino_total,
                treino.T], axis = 0, ignore_index = True
            )
            objetivo_total = pd.concat([
                objetivo_total,
            pd.DataFrame([objetivo])], axis = 0, ignore_index = True)

#treino_total.to_json('treino_total_sc.json')
#objetivo_total.to_json('objetivo_total_sc.json')
```
