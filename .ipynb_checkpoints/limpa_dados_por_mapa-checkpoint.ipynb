```python
import pandas as pd, numpy as np, re
```

```python
def adicionar_lados(df, column):
    # # dada uma coluna 'column' do dataframe 'df', adiciona ao dataframe as colunas 'column'-t e 'column'-ct identicas a 'column'
    # # e ordena as colunas de forma que 'column' é sempre seguida de 'column-t' e 'column-ct'
    # # exemplo: se df possui colunas ['Rating', 'ACS'],
    # # adicionar_lados(df,'rating') retorna df com colunas ['Rating', 'Rating-t', 'Rating-ct', 'ACS']
    
    if not (column in df.columns):
        print('A coluna não existe no dataframe')
        return
    
    temp = df[[column]].copy()
    df = pd.concat( [df, temp.rename(columns = {column : column+'-t'}) ] , axis = 1)
    df = pd.concat( [df, temp.rename(columns = {column : column+'-ct'}) ] , axis = 1)
    del temp
    
    colunas_old = df.columns.tolist()
    r = colunas_old.index(column)
    colunas_new = colunas_old[0:r+1] + colunas_old[-2:] + colunas_old[r+1:-2]
    df = df[colunas_new]
    
    return df
```

```python
df = pd.read_csv('todos_os_dados_final.csv', low_memory = False)
```

```python
df = df.drop(columns = ['Unnamed: 0.3','Unnamed: 0.2', 'Unnamed: 0.1', 'Unnamed: 1', 'Map 1Unnamed: 1', 'Map 2Unnamed: 1',
                       'Map 3Unnamed: 1', 'Map 4Unnamed: 1', 'Map 5Unnamed: 1'])

df = df.rename(columns = {'Unnamed: 0' : 'Player-name', 'Map 1Unnamed: 0' : 'Map 1Player-name',
                          'Map 2Unnamed: 0' : 'Map 2Player-name', 'Map 3Unnamed: 0' : 'Map 3Player-name',
                          'Map 4Unnamed: 0' : 'Map 4Player-name', 'Map 5Unnamed: 0' : 'Map 5Player-name',
                         '+/–' : 'K-D', 'Map 1+/–': 'Map 1K-D', 'Map 2+/–': 'Map 2K-D', 
                         'Map 3+/–': 'Map 3K-D', 'Map 4+/–': 'Map 4K-D', 'Map 5+/–': 'Map 5K-D',
                         '+/–.1' : 'FK-FD', 'Map 1+/–.1': 'Map 1FK-FD', 'Map 2+/–.1': 'Map 2FK-FD', 
                         'Map 3+/–.1': 'Map 3FK-FD', 'Map 4+/–.1': 'Map 4FK-FD', 'Map 5+/–.1': 'Map 5FK-FD'})
```

```python
lista1 = ['R', 'ACS', 'K', 'A', 'ADR', 'FK', 'FD']
lista2 = ['D']
lista3 = ['KAST', 'HS%']
lista4 = ['K-D', 'FK-FD']
listas = [lista1, lista2, lista3, lista4]

temp_4 = []
for lista in listas:
    temp = []
    for elemento in lista:
        temp.append(elemento)
        temp.append('Map 1'+elemento)
        temp.append('Map 2'+elemento)
        temp.append('Map 3'+elemento)
        temp.append('Map 4'+elemento)
        temp.append('Map 5'+elemento)
    temp_4.append(temp)
listas = temp_4
del temp_4

regex1 = '\d+[.]\d+|\d+'
regex2 = '\d+'
regex3 = '\d+(?=%)'
regex4 = '(?<=\+)\w+|\d+[.]\d+|[-]\d+[.]\d+|[-]\d+|\d+'
regexs = [regex1, regex2, regex3, regex4]
```

```python
for i in range(len(listas)):
    for coluna in listas[i]:
        df = adicionar_lados(df, coluna)
```

```python
for i in range(len(listas)):
    for coluna in listas[i]:
        col = []
        col_t = []
        col_ct = []
        for elemento in df[coluna].astype('str'):
            novos_elementos = re.findall(regexs[i], elemento)
            if len(novos_elementos) == 3:
                col.append(float(novos_elementos[0]))
                col_t.append(float(novos_elementos[1]))
                col_ct.append(float(novos_elementos[2]))
            elif len(novos_elementos) == 1 or len(novos_elementos) == 2:
                col.append(float(novos_elementos[0]))
                col_t.append(np.nan)
                col_ct.append(np.nan)
            else:
                col.append(np.nan)
                col_t.append(np.nan)
                col_ct.append(np.nan)
        df[coluna] = pd.Series(col)
        df[coluna+'-t'] = pd.Series(col_t)
        df[coluna+'-ct'] = pd.Series(col_ct)
```

```python
# # transforma KAST e HS% em porcentagens, que atualmente estão de 0 a 100:
for coluna in listas[2]:
    df[coluna] = df[coluna]/100
    df[coluna+'-t'] = df[coluna+'-t']/100
    df[coluna+'-ct'] = df[coluna+'-ct']/100
```

```python
# cria uma coluna booleana 'separador', que é True sempre que a linha é preenchida por NaN, usadas para separar partidas
df = pd.concat([df, df.isnull().apply(lambda x: all(x), axis=1)], axis = 1)
df = df.rename(columns = {0 : 'separador'})

# # recupera os índices das linhas que separam as partidas
separa = df[df['separador'] == True].index.tolist()
separa = [-1] + separa
```

```python
resultado = df[['scores_team_1', 'scores_team_2']]
#df = df.drop(columns = ['scores_team_1', 'scores_team_2'])

resultado = pd.concat([resultado, pd.Series(np.empty(resultado.shape[0], dtype=float))], axis = 1)
resultado = resultado.rename(columns = {0 : 'vitoria_team_1'})
```

```python
for i in range(len(separa)-1):
    if resultado.iloc[separa[i]+1,0] > resultado.iloc[separa[i]+1,1]:
        resultado.iloc[separa[i]+1:separa[i+1],2] = 1
    elif resultado.iloc[separa[i]+1,1] > resultado.iloc[separa[i]+1,0]:
        resultado.iloc[separa[i]+1:separa[i+1],2] = -1
```

```python
# # teste rápido de uma regressão logística
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

colunas_testadas = ['K-D', 'FK-FD']
falso = []
temp_x = []
for j in range(len(colunas_testadas)):
    falso.append(False)
    for k in range(10):
        temp_x.append([0])
falso = tuple(falso)
temp_x = np.array(temp_x)
print(temp_x.shape)
temp_y = []
for i in range(len(separa)-1):
    if separa[i+1] - separa[i] == 11:
        
        temp = df[colunas_testadas].iloc[separa[i]+1:separa[i+1]]
        
        try:
            if temp.isnull().value_counts()[falso] == 10 and resultado.iloc[separa[i]+1,2] != 0:
                temp = np.resize(np.array(temp.transpose()), (10*len(colunas_testadas),1))
                temp_x = np.append(temp_x, temp, axis = 1)
                temp_y.append(resultado.iloc[separa[i]+1,2])
        except:
            pass
del temp
temp_x = np.delete(temp_x, 0, 1)        
X = np.transpose(temp_x)
y = np.array(temp_y)
del temp_x, temp_y
```

```python
scaler = preprocessing.StandardScaler().fit(X)
X = scaler.transform(X)
```

```python

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=2050)

logisticRegr = LogisticRegression()

logisticRegr.fit(x_train, y_train)

predictions = logisticRegr.predict(x_test)

score = logisticRegr.score(x_test, y_test)
print(score)
```

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, predictions)
print(cm)
```
