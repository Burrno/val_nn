```python
import seaborn as sns, numpy as np, matplotlib.pyplot as plt

results = []
results2 = []
results3 = []
array = np.arange(0,1.01,0.02)
for prob in array:
    s = 0
    s2 = 0
    r = 0
    trials = 10000
    md_ = 3
    for i in range(trials):
        g = np.array([0,0,0])
        p = np.array([0,0,0])
        for map_ in range(md_):
            win1 = np.random.rand()
            result = np.random.rand()
            if win1 > 0.5:
                g[map_] = 1
            if result < prob:
                p[map_] = g[map_]
                r+=1
            else:
                p[map_] = int(not(g[map_]))
        done = 0 
        
        mu2 = g.mean()
        if mu2 > 0.5 and p[0]==1:
            s2+=1
        elif mu2<0.5 and p[0]==0:
            s2+=1

        mu = p.mean()
        if g[0] == g[1]:
            if g[0] == 1 and mu > 0.5:
                s += 1
            elif g[0] == 0 and mu < 0.5:
                s+=1
        else:
            if mu2 > 0.5 and mu> 0.5:
                s+=1
            elif  mu2 < 0.5 and mu < 0.5:
                s+=1
    
    results.append(s/trials)
    results2.append(r/(trials*md_))
    results3.append(s2/trials)

plt.figure(figsize = (10,10), facecolor = 'gray')
sns.lineplot(x=array, y = results, color = 'green')
sns.lineplot(x=array, y = results2, color = 'lightblue')
sns.lineplot(x=array, y = results3, color = 'cyan')
plt.vlines([0.62], ymin = 0, ymax = 1)
plt.hlines([0.62], xmin = 0, xmax = 1, color = 'orange')
plt.show()
```
