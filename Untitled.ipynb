```python
import pandas as pd, numpy as np, seaborn as sns, random

from scipy.stats import norm, uniform

df = pd.read_csv('easy_testing.csv')
df.drop(columns = ['Unnamed: 0'], inplace=True)

#df = df.drop([16087,16088,16089,16090,16091,16092]) #df = df.drop([42151,42152,42153,42154,42155,42156])
#df = df.drop([43496,43497,43498,43499,43500,43501])

df['Team'].iloc[36674:36679] = 'VivoKeyd'            # Vivo keyd e Team Vikins ambos são 'VKS', e jogaram um contra o outro 

#df = df.reset_index(drop = True)

```

```python
def separador(df):
    # As partidas estão separadas por linhas "NaN", isto cria uma coluna dizendo se a linha separa uma partida ou não
    temp = df.isnull().apply(lambda x: all(x), axis=1)
    #df = pd.concat([df, ], axis = 1)

    # # recupera os índices das linhas que separam as partidas para usar em laços
    separa = temp[temp == True].index.tolist()
#   separa = [-1] + separa               # garante que pega a primeira partida se a primeira linha não é NaN
    separa = separa + [len(temp)]        # garante que pega a  última  partida se a  última  linha não é NaN
    return separa
```

```python
def RatingJogador(df, separa, função = 'nenhuma'):
    retorno = pd.DataFrame()
    for i in range(len(separa)-1):        
        try:
            ratings = df.loc[separa[i]+1:separa[i+1]-1, ['R','Team','Player-name']]

            times = ratings['Team'].unique()

            time0 = ratings['R'][ratings['Team'] == times[0]]
            time1 = ratings['R'][ratings['Team'] == times[1]]

            #geom = (time1.max()*time2.max())**(1/2)
            #std1 = time1.std()
            #std2 = time2.std()
            #print('std1: ', media1*std1, geom, media1, time1.max(), time1.min())
            #print('std2: ', media2*std2, geom, media2, time2.max(), time2.min())
            
            if função == 'media':
                media0 = time0.mean()
                media1 = time1.mean()
                
                temp1 = ratings['R'][ratings['Team'] == times[0]]*media1
                temp4 = ratings['R'][ratings['Team'] == times[1]]*media0   
                
            else:
                temp1 = ratings['R'][ratings['Team'] == times[0]]
                temp4 = ratings['R'][ratings['Team'] == times[1]] 
                
            temp2 = ratings['Player-name'][ratings['Team'] == times[0]]
            temp3 = pd.concat([temp1, temp2], ignore_index = True, axis = 1)
    
            temp5 = ratings['Player-name'][ratings['Team'] == times[1]]
            temp6 = pd.concat([temp4, temp5], ignore_index = True, axis = 1)

            temp = pd.concat([temp3, temp6], ignore_index=True, axis = 0)

            assert(~(temp[0].isnull().any()))

            retorno = pd.concat([retorno, temp], ignore_index = True, axis = 0)

        except:
            print('partida pulada:\n', ratings)
            pass

    retorno.rename(columns = {0:'R', 1:'Player-name'}, inplace= True)
    return retorno
```

```python
separa = separador(df)
R_pred = RatingJogador(df, separa)
R_pred
```

```python
partidas_olhadas = 10
prep = R_pred.groupby('Player-name').count()> (partidas_olhadas - 1)
```

```python
from sklearn.model_selection import train_test_split
#from sklearn.preprocessing import StandardScaler
#scaler = StandardScaler()

state = random.randint(0,2000)
print(state)

#scaler.fit(X)
#X = scaler.transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=state)


#scaler.fit(X_train)
#X_train = scaler.transform(X_train)
#X_test = scaler.transform(X_test) 
```

```python
from sklearn.neural_network import MLPRegressor

for alph in 10.0 ** -np.arange(0, 7):
        
    regr = MLPRegressor(random_state= abs(state-53),
                        solver = 'sgd',
                        learning_rate = 'adaptive',
                        max_iter=10000,
                        learning_rate_init = 0.1,
                        hidden_layer_sizes=[50,50],
                        activation = 'logistic',
                        alpha = alph,
                        tol = 1e-5,
                        n_iter_no_change = 25
                       )
    #regr.fit(X,y)
    regr.fit(X_train, y_train)
    print('Para alpha =', alph)
    print( 'teste: ', regr.score(X_test, y_test), 'treino: ', regr.score(X_train, y_train), '\n')
```
