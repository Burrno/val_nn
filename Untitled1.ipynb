```python
def separador(df):
    # As partidas estão separadas por linhas "NaN", isto cria uma coluna dizendo se a linha separa uma partida ou não
    temp = df.isnull().apply(lambda x: all(x), axis=1)
    #df = pd.concat([df, ], axis = 1)

    # # recupera os índices das linhas que separam as partidas para usar em laços
    separa = temp[temp == True].index.tolist()
    separa = [-1] + separa               # garante que pega a primeira partida se a primeira linha não é NaN
    #separa = separa + [len(temp)]       # garante que pega a  última  partida se a  última  linha não é NaN
    return separa
```

```python
def correlation_ratio(categories, values):
    mean_v = values.mean()
    categories = np.array(categories)
    values = np.array(values)
    
    ssw = 0
    ssb = 0
    for category in set(categories):
        subgroup = values[np.where(categories == category)[0]]
        
        if len(subgroup) > 10:
            ssw += sum((subgroup-np.mean(subgroup))**2)
            ssb += len(subgroup)*(np.mean(subgroup)-mean_v)**2
    
    print('ssb: ', ssb, '\nssw: ', ssw)
    return (ssb / (ssb + ssw))**.5

colunas = df.columns.tolist()
dict = {}

for coluna in colunas[:-3]:

    coef = correlation_ratio(df['Player-name'], df[coluna])
    dict[coluna] = coef
    print('Eta_squared: {:.4f}\nEta: {:.4f}'.format(coef**2, coef))
    print(colunas[:-3].index(coluna))
```

```python
correlations = pd.DataFrame.from_dict(dict, orient = 'index')
plt.figure(figsize = (20,15))
axes = sns.lineplot(x = correlations.index.tolist(), y = correlations[0], color = 'lightblue')
axes.tick_params(axis = 'x', labelrotation = 90)
axes.axvline(x = 'FD', color = 'black', linewidth = 0.5)

```
